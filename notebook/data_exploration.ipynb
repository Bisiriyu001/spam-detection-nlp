```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# os module is used for file path operations
import os
```

```python
# List all files in the raw data directory to check available datasets
os.listdir("../data/raw")
```

```python
df = pd.read_csv("../data/raw/spam.csv", encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
df = df.iloc[:, :2]
df.columns = ["label", "message"]
df.head()
```

```python
df.isnull().sum()
```

```python
df["label"].value_counts()
```

```python
sns.countplot(x=df["label"])
plt.title("Spam vs Ham Distribution")
plt.show()
```

```python
df["length"] = df["message"].apply(len)
df["length"].describe()
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x="length", hue="label", bins=50)
plt.title("Message Length Distribution")
plt.show()
```
